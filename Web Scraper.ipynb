```python
# This file finds a table w/ the given wiki all time grossing films url and then creates a pandas dataframe with it.
# This is good to review for data acquisition or collection

# Import Necessary modules
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Get the web page
page = requests.get("https://en.wikipedia.org/wiki/List_of_highest-grossing_films#:~:text=Highest-grossing%20films%20%20%20%20Rank%20%20,%20%202015%20%2039%20more%20rows%20")

#Extract HTML
soup = BeautifulSoup(page.content, 'html.parser')

#select just the first table element in the html 
table = soup.find('table', {'class':'wikitable sortable plainrowheaders'})
print(table)
```

```python
# Create an empty list and find all of the th tags and select them
cols = []
for name in table.select('th'):
    
    #this adds only the desired text to a list
    cols.append(name.text.strip())
    
print(cols)

```

```python
# Movie titles we will add to the dataframe later on
titles = cols[6:]
```

```python
# Same as the steps above
gross = []
for film in table.select('td'):
    gross.append(film.text.strip())
print(gross)
```

```python
# This deletes every 5th element in the list. those are the references and we dont want those
del gross[4::5]
print(gross)
```

```python
# Turn into a numpy array and reshape the data so that it is in a matrix
import numpy as np
gross = np.array(gross)
mat = np.reshape(gross, (50,4))
print(mat)
```

```python
# Convert titles to a np array and create the dataframe
titles = np.array(titles)
df = pd.DataFrame(mat, columns = ["Rank","Peak", "Amount Grossed in USD", "Year"])
```

```python
# Add the titles to the df
df['Title'] = titles
print(df)
```
